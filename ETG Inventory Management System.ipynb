#### Login

```python
def login(i_username, i_password):
    with open('Shop.json', 'r') as txt:
        record = json.load(txt)
        credentials = record['credentials']
        username = credentials['Shop Name']
        password = credentials['Password']

        if i_username != username or i_password != password:
            return False
        else:
            return True
```

#### add_product

```python
def add_product():
    product_id = (input("Enter product ID: ").strip())
    product_name = input('Enter product name: ').strip()
    stock = int(input('Enter stock: ').strip())
    quantity = input('Enter quqntity: ').strip()
    MFD = input('Enter MFD: ').strip()
    price = float(input('Enter price: ').strip())

    if not product_name:
        print('Please fill all the fields correctly ...')
        return add_product()

    with open('Shop.json', 'r') as txt:
        record = json.load(txt)
        record['Products'][product_id] = {
            "Name": product_name,
            "Stock": stock,
            "Quantity" : quantity,
            "MFD" : MFD,
            "Price" : price
        }
        with open('Shop.json', 'w') as txt:
            json.dump(record, txt, indent=2, sort_keys=True)
           
    return Shopkeeper_panel()
```

#### update_product

```python
def update_product():
    product_id = input('Enter product id: ').strip()
    if not product_id:
        print("Please enter a valid id ...")
        return update_product()

    with open('Shop.json', 'r') as txt:
        record = json.load(txt)
        ids = record['Products'].keys()
        if product_id not in ids:
            print("Please enter a valid id ...")
            return update_product()

        u_name = input('Enter updated product name: ').strip()
        u_stock =input('Enter updated stock: ').strip()
        u_quantity = input('Enter updated quantity: ').strip()
        u_MFD = input('Enter updated MFD: ').strip()
        u_price = input('Enter updated price: ').strip()
        

        if not u_name:
            u_name = record['Products'][product_id]['Name']
            
        if not u_stock:
            u_stock = record['Products'][product_id]['Stock']
        
        if not u_quantity:
            u_quantity = record['Products'][product_id]['Quantity']

        if not u_MFD:
            u_MFD = record['Products'][product_id]['MFD']
            
        if not u_price:
            u_price = record['Products'][product_id]['Price']
            
        updated_product = {
            "Name": u_name,
            "Stock": int(u_stock),
            "Quantity" : u_quantity,
            "MFD" : u_MFD,
            "Price" : int(u_price)
        }

        record['Products'][product_id] = updated_product

        with open('Shop.json', 'w') as txt:
            json.dump(record, txt,indent=2, sort_keys=True)
            
    return Shopkeeper_panel()
```

#### product_list

```python
def product_list():
    with open('Shop.json','r') as txt:
        record = json.load(txt)
        txt.close()
        for product_id, product in record['Products'].items():
            print(f"Product ID: {product_id} ",f": {product['Name']}")
    return Customer_panel()
```

#### delete_product

```python
def delete_product():
    product_id = input('Enter product id: ').strip()
    if not product_id:
        print("Please enter a valid id ...")
        return update_product()

    with open('Shop.json', 'r') as txt:
        record = json.load(txt)
        ids = record['Products'].keys()
        if product_id not in ids:
            print("Please enter a valid id ...")
            return Shopkeeper_panel()

        record['Products'].pop(product_id)

        with open('Shop.json', 'w') as txt:
            json.dump(record, txt,indent=2, sort_keys=True)
        
            
    return Shopkeeper_panel()
```

#### search_product

```python
def search_product():
    product_name = input('\nSearch product (Enter the product name): ').strip()
    with open('Shop.json', 'r') as txt:
        record = json.load(txt)
        for product_id, product in record['Products'].items():
            if product['Name'].find(product_name) > -1:
                print(f"Product ID: {product_id}")
                print(f"Product Name: {product['Name']}")
                print(f"Stock: {product['Stock']}")
                print(f"Quantity: {product['Quantity']}")
                print(f"MFD: {product['MFD']}")
                print(f"Price: {product['Price']}")
                print("-"*10)
            
        return Customer_panel()
```

#### transaction

```python
def transaction(pdt_ids) :
    #import time
    time.ctime()
    #import json
    #sales = {}
    report=[]
    
    total=0

    with open('Shop.json','r') as j_f:
        record = json.load(j_f)
            
            
    #import json
    #a_file = open("Sales.json", "r")
    #json_object = json.load(a_file)
    #a_file.close()
            
    order={}

            
    for pdt_id in pdt_ids:
        print(f"Selected Product Name: {record['Products'][pdt_id]['Name']} -",f"Price : {record['Products'][pdt_id]['Price']}")
        order[pdt_id]=input("Enter the No. of quqntity you want: ")
        
    print("\n**************Bill*****************\n")
    print("Receipt No :\n")
    print(time.ctime(),'\n')
    for x, y in order.items():
        print(f"{record['Products'][x]['Name']}  : {record['Products'][x]['Price']} *",y,"=",int(record['Products'][x]['Price'])*int(y))
        total+= (int(record['Products'][x]['Price'])*int(y))
        
    for k, v in order.items():
    
        #updating product data in json file
        record ['Products'][k]['Stock'] = int(int(record ['Products'][k]['Stock']) - int(v))
        j_f = open("Shop.json", "w+")
        json.dump(record, j_f)
#       j_f.close()

        #generating data for sales file    
        report.append( '\n\n'+str(time.ctime())+'\n'+'Product name: '+ record['Products'][k]['Name']+"\nRemaining Stock "+str(record['Products'][k]['Stock'])+"\nAmt of one piece: "+ str(record['Products'][k]['Price'])+ "\nTotal Billing: "+str(total))
        fdj = open('Sales.json','a')
        fdt = open('Sales.txt','a')
        for i in report:
            fdj.write(i)
        fdj.close()
        
        for i in report:
            fdt.write(i)
        fdt.close()
            
    print("------------------------------------")
    print("Billing Amount =",total,"Rs")
    print("************************************")

        
    #with open("Sales.json", "w") as outfile:
        #json.dump(Sales, outfile, indent=1)
        
    return Customer_panel()
```
